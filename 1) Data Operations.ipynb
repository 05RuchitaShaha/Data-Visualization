```python
import pandas as pd
```

```python
data = {
    "Name" : ["John", "Alice", "Bob", "Emma", "Mike"],
    "Age" : [25,30,28,32,27],
    "Salary" : [50000, 60000, 55000, 70000, 52000],
    "Department" : ["HR", "Marketing", "HR", "Finance", "Marketing"]
}
```

```python
df = pd.DataFrame(data)
df
```

### @@@@

'SELECT – choose specific columns from a dataset while excluding others'

```python
## Select specific Columns

list_col = ['Age', 'Salary']
df[list_col]
```

### @@@@

'FILTER - Simple vs Complex : allows to extract specific rows from dataset based on certain condition.
Simple Filter : selecting all rows based on condition
Complex Filter : select rows based on multiple conditions'

```python
filtered_data = df[df["Age"] > 30]
filtered_data
```

```python
df["Department"] == "Marketing"     # one df for Conditional
```

```python
df[df["Department"] == "Marketing"]        # another df for Selection
```

```python
complex_filter = df[(df["Department"] == "Marketing") & (df['Salary'] > 50000)]
complex_filter
```

### @@@@

'SORT - arrange rows of dataset in a specific order based on values in 1 or more cols'

```python
sort_data_asc = df.sort_values(by = "Salary")
sort_data_asc
```

```python
sort_data_des = df.sort_values(by = "Salary", ascending = False)
sort_data_des
```

### @@@@

'GROUP & AGGREGATE - Grouping is categorizing data based on values in 1 or more cols
after grping perform aggregate functions like SUM, AVG, COUNT, etc '

```python
grouped_data = df.groupby("Department").agg({'Salary' : "min"})
grouped_data
```

```python
grouped_data = df.groupby("Department").agg({'Salary' : "mean"})
grouped_data
```

```python
grouped_data = df.groupby("Department").agg({'Salary' : "sum"})
grouped_data
```

### @@@@

'MERGE - combines 2 or more DataFrames based on common cols or indices.
Brings data together from diff sources & create single DF with relevant info.'

```python
data1 = {
    "EmpID" : [1,2,3],
    "Name" : ["John", "Alice", "Bob"],
    "Age" : [25,30,28]
}

data2 = {
    "ID" : [1,2,4],
    "Salary" : [50000, 6000, 65000],
    "Department" : ["HR", "Marketing", "HR"],   
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2           ## ignore the indexes returned by python by default
```

```python
'INNER JOIN'
# Merge based on "ID" coln ( when both dfs had ID common)

merged_data = pd.merge(df1, df2, on = "ID")      
merged_data
```

```python
'INNER JOIN'
# # Merge based on "ID" & EmpID coln ( when both dfs have diff matching cols)

merged_data = pd.merge(df1, df2, left_on = ['EmpID'], right_on = ["ID"], how = 'inner')      
merged_data
```

```python
'OUTER JOIN'
# # Merge based on "ID" & EmpID coln ( when both dfs have diff matching cols)

merged_data = pd.merge(df1, df2, left_on = ['EmpID'], right_on = ["ID"], how = 'outer')      
merged_data
```

```python
'LEFT OUTER JOIN'
# # Merge based on "ID" & EmpID coln ( when both dfs have diff matching cols)

merged_data = pd.merge(df1, df2, left_on = ['EmpID'], right_on = ["ID"], how = 'left')      
merged_data
```

```python
'RIGHT OUTER JOIN'
# # Merge based on "ID" & EmpID coln ( when both dfs have diff matching cols)

merged_data = pd.merge(df1, df2, left_on = ['EmpID'], right_on = ["ID"], how = 'right')      
merged_data
```

### @@@@

'PIVOT - Reshapes the data, converts unique values of 1 coln into multiple cols
Useful for summarizing & presenting data in a more structured form'

```python
data = {
    "Name" : ["John", "Alice", "Bob", "John", "Alice", "Bob"],
    "Month" : ["Jan", "Jan", "Jan", "Feb", "Feb", "Feb"],
    "Sales" : [100, 150, 120, 200, 180, 170]
}
df = pd.DataFrame(data)
df
```

```python
pivot_data = pd.DataFrame(df.pivot(index = "Name", columns = "Month", values = "Sales" ))
pivot_data

## Nested Data Format
```

```python
data3 = {
    "Name" : ["John", "Alice", "Bob", "John", "Alice", "Bob", "Bob"],
    "Month" : ["Jan", "Jan", "Jan", "Feb", "Feb", "Feb", "Jan"],
    "Sales" : [100, 150, 120, 200, 180, 170, 40]
}
df3 = pd.DataFrame(data3)
df3

# when same person, same name, diff sales
```

```python
group_Bob = df3.groupby("Month" and "Name").agg({"Sales" : "sum"})
group_Bob
```

```python
pivot = pd.DataFrame(df3.pivot(index = "Name", columns = "Month", values = "Sales" ))
pivot
```

### @@@@

'UNPIVOT - opposite of pivot, unfolding pivot
data transformed from multiple cols back to longer format
helpful when to analyze data in a Tabular Format'

```python
data4 = {
    "Name" : ["John", "Alice", "Bob"],
    "Jan" : [100, 150, 170],
    "Feb" : [200, 180, 170]
}

df4 = pd.DataFrame(data4)
df4
```

```python
unpivot = pd.melt(df3, id_vars = ["Name"], var_name = "Month", value_name = "Sales")
unpivot
```

```python
unpivot = pd.melt(df4, id_vars = ["Name"], var_name = "Month", value_name = "Sales")
unpivot
```

### @@@@

'WINDOWING - allows to perform calculations on a Subset of data, defined by window or frame,
rather than entire dataset.
Computes Rolling Averages, Cumulative Sums, & other Aggregations on 
sliding windows of data'

```python
# for stock market data used 

data5 = {
    "Month" : ["Jan", "Feb", "Mar", "April", "May", "June"],
    "Sales" : [100, 120, 80, 150, 130, 110]
}

df5 = pd.DataFrame(data5)
df5
```

```python
# 3 month rolling avg

df5["Rolling_Avg"] = df5["Sales"].rolling(window=3).mean()
df5
```

```python
df5["Rolling_Avg"] = df5["Sales"].rolling(window=3).max()
df5
```

```python
df5["Rolling_Avg"] = df5["Sales"].rolling(window=3).count()
df5
```
