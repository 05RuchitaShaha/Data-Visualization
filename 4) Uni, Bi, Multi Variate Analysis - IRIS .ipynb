### Create Dummy Data for Missing Values

```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
dic = {"name" : ["Ram", "Samena", "Teja", "Rahul", "Sanjay", "Hari"],
      "age" : [24, 67, np.nan, 34, 19, np.nan],
      "qualification" : ["BE", np.nan, "BSc", "PhD", "Data Science", np.nan],
      "salary" : [45000, 76000, 12000, np.nan, 23000, 98000]}

dfn = pd.DataFrame(dic)
dfn
```

### Simple Imputer

```python
dic = {"name" : ["Ram", "Samena", "Teja", "Rahul", "Sanjay", "Hari"],
      "age" : [24, 67, np.nan, 34, 19, np.nan],
      "qualification" : ["BE", np.nan, "BSc", "PhD", "Data Science", np.nan],
      "salary" : [45000, 76000, 12000, np.nan, 23000, 98000]}

dfn = pd.DataFrame(dic)
dfn
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = "mean")
si

# set strategy type as 'mean', 'median', 'most_frequent', 'constant' etc to replace missing values
```

```python
dfn["age"] = si.fit_transform(dfn[["age"]])
dfn
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = "most_frequent")
si
```

```python
dfn["salary"] = si.fit_transform(dfn[["salary"]])
dfn
```
